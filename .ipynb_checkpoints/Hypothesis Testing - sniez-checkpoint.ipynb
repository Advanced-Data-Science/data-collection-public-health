```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns

df = pd.read_excel("../Preprocessing/FDAcleaned.xlsx")
```

The hypothesis being explored here is that the distribution of product types among recalls has not changed significantly over time, aside from random fluctuations.


Resources for creating a chi-square test in python  
https://www.statology.org/pandas-crosstab-aggfunc/  
https://stackoverflow.com/questions/24767161/can-we-generate-contingency-table-for-chisquare-test-using-python

```python
# chi-square test

# convert the date to a datetime object and isolate the year
df['recall_date'] = pd.to_datetime(df['recall_date'], format="%m/%d/%Y")
df['year'] = df['recall_date'].dt.year

# 2014 and 2025 data is only partial, keeping it would erroneously inflate the differences from expected value
df_processed = df[~df['year'].isin([2014, 2025])]

# remove Tobacco since n=4 violates chi-square assumptions
df_processed = df_processed[df_processed['product_type'] != 'Tobacco']

# create a contingency table to base the chi square test on
# aggregate using unique event ID's
ct = pd.crosstab(df_processed['product_type'], df_processed['year'], values=df_processed['event_id'], aggfunc=pd.Series.nunique).fillna(0)

# run the chi-square test
c, p, dof, expected = chi2_contingency(ct) 
# note: the expected variable returns an array of expected values that we can use for a heatmap

# print the test stat
print(f"Test statistic: {c:.2f}")
# print the p-value
print(f"p = {p:.3f}")
# print degrees of freedom
print(f"df = {dof}")

# cramer's v
n = ct.values.sum()
r, c, = ct.shape # tuple unpacking: r=rows, c=columns
cramers_v = np.sqrt(c / (n * (min(r-1, c-1)))) # calculate cramer's v
print(f"Cramer's V = {cramers_v:.3f}")
```

To get a better idea of where the most variation is coming from, we create a heatmap of the residuals  

Additional resources:  
https://www.youtube.com/watch?v=Jb9sxlnhZMo  
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html


```python
# create the dataframe of expected values as calculated in the chi-square test
expected_df = pd.DataFrame(expected, index=ct.index, columns=ct.columns)

# calculate the standardized residuals using the contingency table and expected values table
residuals = (ct - expected_df) / np.sqrt(expected_df)

# create a heatmap
plt.figure(figsize=(12,8))
sns.heatmap(residuals, annot=True, fmt=".2f", cmap="Spectral")
plt.title("Standardized Residuals")
plt.xlabel("Year")
plt.ylabel("Product Type")
plt.show()
```
